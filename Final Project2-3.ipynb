```python
def influencers(x):
    url= x
    results=requests.get(url)
    soup = BeautifulSoup(results.content, 'html.parser')
    title=[]
    Totalfollowers = [item.text for item in soup.select('.td:nth-of-type(4):not([id])')]
    followerspain = [item.text for item in soup.select('.td:nth-of-type(5):not([id])')]
    authenticengament=[item.text for item in soup.select('.td:nth-of-type(6):not([id])')]
    Engagementavg=[item.text for item in soup.select('.td:nth-of-type(7):not([id])')]
    len_songs=len(soup.select('td[style="width:280px;min-width:280px;max-width:280px;"]'))
    account2=[]
    links=[]
    for i in range(len_songs):
        title.append(soup.select('td[style="width:280px;min-width:280px;max-width:280px;"]'))
        tatd=str(title)
        tatd=tatd.replace('<td class="td" data-v-1c693a69="" style="width:280px;min-width:280px;max-width:280px;">','')
        tatd=tatd.split(',')
        df = pd.DataFrame (tatd, columns = ['column_name'])
        df[['Name','Account']] = df.column_name.str.split("@",expand=True,)
        df=df.drop(columns=['column_name'])
        df = df.drop_duplicates(subset='Name', keep="first")
        df = df.drop_duplicates(subset='Account', keep="first")
        df.reset_index()
        followers = pd.Series(Totalfollowers)
        followerspain=pd.Series(followerspain)
        df['Totalfollowers'] = followers
        df['followerspain']=followerspain
        authenticengament = pd.Series(authenticengament)
        df['authenticengament']=authenticengament.values
        account=list(df['Account'])
        for i in account:
            account=i.replace('</td>','')
            account2.append(account)
        df['Account2']= account2
        for i in account2: 
            links.append('https://www.instagram.com/'+i+'/?hl=es')
        df['links']=links

        return df
```
