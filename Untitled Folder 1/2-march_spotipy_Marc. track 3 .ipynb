```python
def get_all(user_id, playlist_id):
    results = sp.user_playlist_tracks(user_id, playlist_id)
    tracks = results ['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    id2=[]
    artist=[]
    song=[]
    for i in range(len(tracks)):
        song.append(tracks[i]['track']['name'])
        artist.append(tracks[i]['track']['artists'][0]['name'])
        id2.append(tracks[i]['track']['id'])
    dict = {id2[i]: (song[i], artist[i]) for i in range(len(id2))}
    audio_feats = []
    for x in tqdm(range(len(tracks))):
        audio_feats.extend(sp.audio_features(tracks=tracks[x]['track']['id']))
    df_name = pd.DataFrame(audio_feats)
    return df_name
```

```python
def songsinfo(user_id, playlist_id):
    results = sp.user_playlist_tracks(user_id, playlist_id)
    tracks = results ['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    id2=[]
    artist=[]
    song=[]
    for i in range(len(tracks)):
        song.append(tracks[i]['track']['name'])
        artist.append(tracks[i]['track']['artists'][0]['name'])
        id2.append(tracks[i]['track']['id'])
    dataframe=pd.DataFrame(list(zip(id2, artist, song)),columns=['ids','names', 'artist'])
    return dataframe 
```

# Finding the right K

```python
allmusic_scaled_df2=allmusic_scaled_df.drop(['cluster_no','id'],axis=1)
```

```python
allmusic_scaled_df2.columns
```

```python
K=range(5,18)
inertia=[]
for k in K:
    kmeans=KMeans(n_clusters=k,random_state=5)
    kmeans.fit(allmusic_scaled_df2)
    inertia.append(kmeans.inertia_)
```

```python
#plot the resulting inertia - elbow 
plt.figure(figsize=(16,8))
plt.plot(K,inertia,'bx-')
plt.xlabel('k=noofclusters')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K),max(K)+1,1.0))
plt.title('elbow method inertia to find optimal K');
```

```python
K=range(5,18)
sil=[]
for k in K:
    kmeans=KMeans(n_clusters=k,random_state=5)
    kmeans.fit(allmusic_scaled_df2)
    sil.append(silhouette_score(allmusic_scaled_df2,kmeans.predict(allmusic_scaled_df2)))
#silhouette score 
# if 1 - the clusters are very separated from eachother and clearly distinguished
# if 0 - clusters are indifferent - the distance between them are not significant and you cannot distinguish
# if -1 clusters are incorrectly assigned 
```

```python
#plot the resulting silhouette 
plt.figure(figsize=(16,8))
plt.plot(K,sil,'bo-')
plt.xlabel('k=noofclusters')
plt.ylabel('silhouette')
plt.xticks(np.arange(min(K),max(K)+1,1.0))
plt.title('silhouette score');
```
